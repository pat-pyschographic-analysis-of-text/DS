```python
import pandas as pd

pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('twitter_psychographics.csv', index_col='user')
display(df.head(), df.shape)
```

```python
df
```

```python
dfg = df.pivot_table(values='raw_score', columns='name', index=df.index)
dfg
```

```python
# Pearson correlation of users
dfg.T.corr()
```

```python
# Heatmap

import seaborn as sns

sns.heatmap(dfg.T.corr(), cmap="Blues")
```

```python
# PCA similarity

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
df_pca = pd.DataFrame(pca.fit_transform(dfg),index=dfg.index,
                      columns=['PC1','PC2'])
df_pca.head()

ax = df_pca.plot(kind='scatter', x='PC2', y='PC1', figsize=(12,8))

for i, name in enumerate(df_pca.index):
    ax.annotate(name, (df_pca.iloc[i].PC2, df_pca.iloc[i].PC1))
```

```python
# t-SNE similarity

from sklearn.manifold import TSNE

tsne = TSNE()
df_tsne = pd.DataFrame(tsne.fit_transform(dfg),index=dfg.index,
                      columns=['C1','C2'])
display(df_tsne)

ax = df_tsne.plot(kind='scatter', x='C2', y='C1', figsize=(12,8))

for i, name in enumerate(df_tsne.index):
    ax.annotate(name, (df_tsne.iloc[i].C2, df_tsne.iloc[i].C1))
```

```python
from sklearn.linear_model import LogisticRegression

X = dfg
y = dfg.index

lr = LogisticRegression()
lr.fit(X,y)
r = pd.DataFrame(lr.predict_proba(X), index=dfg.index)
r
```
